Here is a link to the “Auto MPG” dataset in the UC Irvine data repository: https://archive.ics.uci.edu/ml/datasets/Auto+MPG


```python
# Importing the modules neccessary to complate the tasks in the assignment.
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Step 1
- Load the data in the (yellow highlighted) Data Folder in the file auto-mpg.data into a pandas DataFrame.

```python
auto_mpg = pd.read_fwf('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', header=None, names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Origin', 'Car Name']).set_index('Car Name')
auto_mpg.head()
```

#### Step 2
- Some of the horsepower data is marked as missing with a ‘?’. You need to replace this ‘?’ with a pandas-appropriate missing value, then convert the column to numeric.
- The origin column values are 1, 2, and 3. Convert these values respectively to ‘USA’, ‘Asia’, and ‘Europe’
- Using replace() module to clean up missing values.

```python
auto_mpg = auto_mpg.replace('?', np.nan)
auto_mpg['Origin'].replace(1, 'USA', inplace=True)
auto_mpg['Origin'].replace(2, 'Asia', inplace=True)
auto_mpg['Origin'].replace(3, 'Europe', inplace=True)
auto_mpg.Horsepower = pd.to_numeric(auto_mpg.Horsepower)
auto_mpg.head()
```

#### Step 3
- Create a bar chart that shows the distribution for cylinders.
- Create a scatterplot that shows the relationship between horsepower and weight.

```python
dataset = auto_mpg.groupby('Cylinders')
dataset['Cylinders'].count().plot.bar(title="Cylinder Distribution")
plt.show()
```

#### Step 4
- Create a scatterplot that shows therelationship between horsepower and weight

```python

sns.scatterplot(x = auto_mpg['Horsepower'], y = auto_mpg['Weight'])
```

#### Step 5 
- What is  the correlation between horsepower and acceleration in the “Auto MPG” dataset?


```python
sns.swarmplot(data = auto_mpg,
              x = 'Horsepower', 
              y = 'Acceleration',  
              hue = 'Origin')

plt.title('Correlation between Horsepower and Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
```
